```python
import matplotlib.pyplot as plt
import cv2

# Load your images (OpenCV loads in BGR, so convert to RGB)
img1 = cv2.imread('dataset_phase_2/healthy/test32.jpg', cv2.IMREAD_COLOR_RGB)
img2 = cv2.imread('dataset_phase_2/healthy/healthy47_png.rf.906e18e18ad480373ba9ffda86ac10ca.jpg', cv2.IMREAD_COLOR_RGB)
img3 = cv2.imread('dataset_phase_2/retinoblastoma/11_jpeg.rf.aa900289a9d307d0c371bd641bb54b2d.jpg', cv2.IMREAD_COLOR_RGB)

# Create subplots: 1 row, 3 columns
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot each image
axs[0].imshow(img2)
axs[0].set_title('Око з червоним рефлексом')
axs[0].axis('off')

axs[1].imshow(img1)
axs[1].set_title('Око без червоного рефлексу')
axs[1].axis('off')

axs[2].imshow(img3)
axs[2].set_title('Око з ретинобластомою')
axs[2].axis('off')

# Show the full figure
plt.tight_layout()
plt.show()
```
