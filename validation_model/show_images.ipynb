```python
import matplotlib.pyplot as plt
import cv2

# Load your images (OpenCV loads in BGR, so convert to RGB)
img1 = cv2.imread('processed_dataset/training/eye/eye_45.png', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('processed_dataset/training/eye/eye_143.png', cv2.IMREAD_GRAYSCALE)
img3 = cv2.imread('processed_dataset/training/eye/eye_11.png', cv2.IMREAD_GRAYSCALE)

# Create subplots: 1 row, 3 columns
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot each image
axs[0].imshow(img1, cmap='gray')
axs[0].set_title('Eye 1')
axs[0].axis('off')

axs[1].imshow(img2, cmap='gray')
axs[1].set_title('Eye 2')
axs[1].axis('off')

axs[2].imshow(img3, cmap='gray')
axs[2].set_title('Eye 3')
axs[2].axis('off')

# Show the full figure
plt.tight_layout()
plt.show()
```
