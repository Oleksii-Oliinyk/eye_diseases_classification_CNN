```python
import os
import random
```

```python
from dataset_prep_lib import read_images, process_validation_images, save_images
```

```python
not_eye_read_path = ["unproccesed_dataset/no_eye/bad_eye", "unproccesed_dataset/no_eye/face", "unproccesed_dataset/no_eye/fake_eye", "unproccesed_dataset/no_eye/random"]

eye_read_path = "unproccesed_dataset/eye"
```

```python
unproccesed_not_eye_images = []

for pth in not_eye_read_path:
    arr_images = read_images(pth)
    unproccesed_not_eye_images.append(arr_images)

unproccesed_eye_images = read_images(eye_read_path)
```

```python
dataset_folder = "processed_dataset_foo/"

train_path = dataset_folder + "training/"
validation_path = dataset_folder + "validation/"

dataset_classes = ["eye", "not_eye"]

for dataset_class in dataset_classes:
    os.makedirs(train_path + dataset_class, exist_ok=True)
    os.makedirs(validation_path + dataset_class, exist_ok=True)
```

```python
proccesed_not_eye_images = []

for images in unproccesed_not_eye_images:
    proccesed_not_eye_images.append(process_validation_images(images))

proccesed_eye_images = process_validation_images(unproccesed_eye_images)
```

```python
train_eye_images = []
validation_eye_images = []

random.shuffle(proccesed_eye_images)
split_index = int(len(proccesed_eye_images) * 0.85)
train_eye_images = proccesed_eye_images[:split_index]
validation_eye_images = proccesed_eye_images[split_index:]
```

```python
print(len(train_eye_images))
print(len(validation_eye_images))
save_images(train_eye_images, train_path + dataset_classes[0], "eye")
save_images(validation_eye_images, validation_path + dataset_classes[0], "eye")
```

```python
train_not_eye_images = []
validation_not_eye_images = []
for i in range(len(proccesed_not_eye_images)):
    random.shuffle(proccesed_not_eye_images[i])
    split_index = int(len(proccesed_not_eye_images[i]) * 0.85)
    train = proccesed_not_eye_images[i][:split_index]
    val = proccesed_not_eye_images[i][split_index:]
    train_not_eye_images.append(train)
    validation_not_eye_images.append(val)
```

```python
names_arr = ["bad_eye", "face", "fake_eye", "random_images"]

for i in range(len(names_arr)):
    save_images(train_not_eye_images[i], train_path + dataset_classes[1], names_arr[i])
    save_images(validation_not_eye_images[i], validation_path + dataset_classes[1],  names_arr[i])
```
