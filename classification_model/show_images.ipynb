```python
import matplotlib.pyplot as plt
import cv2
import matplotlib.image as mpimg

# Load your images (OpenCV loads in BGR, so convert to RGB)
img1 = mpimg.imread('processed_dataset_final/training/cataract/cataract_102.png')
img2 = mpimg.imread('processed_dataset_final/training/cataract/cataract_102_flipped.png')
img3 = mpimg.imread('processed_dataset_final/training/cataract/cataract_39_rotated.png')

# Create subplots: 1 row, 3 columns
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot each image
axs[0].imshow(img1)
axs[0].set_title('Original')
axs[0].axis('off')

axs[1].imshow(img2)
axs[1].set_title('Flipped')
axs[1].axis('off')

axs[2].imshow(img3)
axs[2].set_title('Rotated')
axs[2].axis('off')

# Show the full figure
plt.tight_layout()
plt.show()
```
